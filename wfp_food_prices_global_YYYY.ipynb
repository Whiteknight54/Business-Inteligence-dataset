## Load All Uncleaned WFP Data

### Subtask:
Iterate through the `resource_map` to load each yearly `wfp_food_prices_global_YYYY.csv` file into a list of DataFrames, without any filtering or cleaning.


**Reasoning**:
To load all yearly WFP food prices CSV files into a list of DataFrames without any filtering or cleaning, I will implement a loop that iterates through the `resource_map`, constructs file paths, checks for file existence, loads the CSVs into DataFrames, appends them to a list, and includes error handling.



````python
import os
import pandas as pd

raw_yearly_dfs = []

for year, fname in resource_map.items():
    filepath = os.path.join(base_path, fname)

    if not os.path.exists(filepath):
        print(f"⚠️ File not found for {year}: {filepath}")
        continue

    try:
        df_raw_yearly = pd.read_csv(filepath)
        raw_yearly_dfs.append(df_raw_yearly)
        print(f"Loaded raw data for {year}: {len(df_raw_yearly):,} records.")
    except Exception as e:
        print(f"❌ Error loading {filepath}: {e}")

print(f"\n✅ Successfully loaded {len(raw_yearly_dfs)} raw DataFrames.")
````

## Combine All Uncleaned Data

### Subtask:
Concatenate all the loaded yearly DataFrames into a single `raw_master_df` containing all WFP data from 2010 to 2024.


**Reasoning**:
To combine all the loaded yearly DataFrames into a single `raw_master_df`, I will check if the `raw_yearly_dfs` list is not empty and then use `pd.concat` to concatenate them. If the list is empty, an empty DataFrame will be initialized.



````python
if raw_yearly_dfs:
    raw_master_df = pd.concat(raw_yearly_dfs, ignore_index=True)
    print(f"\n✅ Combined all uncleaned WFP data: {len(raw_master_df):,} total records.")
else:
    raw_master_df = pd.DataFrame()
    print("\n⚠️ No raw WFP data was loaded. raw_master_df is empty.")
````

**Reasoning**:
The next step is to export the combined `raw_master_df` to a CSV file in Google Drive, as specified in the subtask instructions.



````python
output_file_name = "wfp_food_prices_global 2010-2024.csv"
raw_output_path = f"/content/drive/MyDrive/{output_file_name}"

# Export the raw_master_df to a CSV file
if not raw_master_df.empty:
    try:
        raw_master_df.to_csv(raw_output_path, index=False)
        print(f"✅ Combined uncleaned WFP data successfully saved to: {raw_output_path}")
    except Exception as e:
        print(f"❌ Error saving the combined uncleaned WFP data to Google Drive: {e}")
else:
    print(f"❌ Cannot save an empty DataFrame. The combined uncleaned WFP data file was not created.")
````
